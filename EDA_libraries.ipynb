# Autoviz

```python
!pip install autoviz
```

```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
```

```python
import seaborn as sns

df4 = sns.load_dataset('titanic')
df4.head()
```

```python
filename = ""
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=df4,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```
